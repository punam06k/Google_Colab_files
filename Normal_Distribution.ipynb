```python
glaxo_df = pd.read_csv("GLAXO.csv")
glaxo_df.head()
```

```python
beml_df = beml_df[['Date','Close']]          # beml_df.iloc[:,[0,5]]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df['Date'] = pd.to_datetime(glaxo_df['Date'])    # change the data type of Date 
beml_df['Date'] = pd.to_datetime(beml_df['Date'])
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
glaxo_date = pd.DatetimeIndex(glaxo_df['Date'])    # Date is column ,crete as index
beml_date =  pd.DatetimeIndex(beml_df['Date'])
```

```python
glaxo_df = glaxo_df.set_index(glaxo_date).drop('Date',axis = 1)     # axis = 1 columns axis = 0 rows
```

```python
glaxo_df
```

```python
beml_df = beml_df.set_index(beml_date).drop('Date',axis = 1)
```

```python
beml_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib incline
plt.plot(glaxo_df.Close);
plt.xlabel('Time');
plt.ylabel('Close Price');
```

```python
plt.plot(beml_df.Close);
plt.xlabel('Time');
plt.ylabel('Close Price');
```

```python
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
```

```python
glaxo_df
```

```python
# drop first row since it is NAN
glaxo_df = glaxo_df.dropna()
beml_df = beml_df.dropna()
```

```python
glaxo_df
```

```python
# plot the gains
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 5));
plt.plot(glaxo_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 5));
plt.plot(beml_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
import seaborn as sn

plt.figure(figsize = (10, 5))
plt.subplot(1,2,1)
plt.suptitle('Distribution',fontsize = 14)

sn.distplot(glaxo_df.gain, label = 'GLAXO');

plt.xlabel('gain')
plt.ylabel('Density');
plt.legend()

plt.subplot(1,2,1)

sn.distplot(beml_df.gain, label = 'BEML');

plt.xlabel('gain')
plt.ylabel('Density');
plt.legend()
```

```python
print('Mean: ', round(glaxo_df.gain.mean(), 4))
print('Standard Deviation: ', round(glaxo_df.gain.std(), 4))
```

```python
print('Mean: ', round(beml_df.gain.mean(), 4))
print('Standard Deviation: ', round(beml_df.gain.std(), 4))
```
